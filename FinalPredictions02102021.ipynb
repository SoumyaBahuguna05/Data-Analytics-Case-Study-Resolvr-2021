```python
products = ['r2m','r3m','r2c','r4c','r4a']
output2 = []
def mape(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.mean(np.abs((actual - pred) / actual)) * 100

#    """SUPPORT VECTOR REGRESSION"""
for i in products:
  
  print("The time series considered is: ",i) 
  x1,x2,x3,y=df_clean[i+'-1'],df_clean[i+'-2'],df_clean[i+'-3'],df_clean[i]
  x1,x2,x3,y=np.array(x1),np.array(x2),np.array(x3),np.array(y)
  x1,x2,x3,y=x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),y.reshape(-1,1)
  final_x=np.concatenate((x1,x2,x3),axis=1)
  #print(final_x)

  from sklearn.preprocessing import StandardScaler
  scx = StandardScaler()
  scy = StandardScaler()
  x = scx.fit_transform(final_x) #fit will calculate the mean and sd and transform applies the formula
  y = scy.fit_transform(y)
  from sklearn.svm import SVR
  lin_model=SVR(kernel = 'rbf')

  lin_model.fit(final_x,y)

  test_predictions = []

  first_eval_batch = final_x[-1:]
  current_batch = first_eval_batch #.reshape((1, 3, 1))
  print(current_batch)
  current_batch.reshape(-1, 1)
  for j in range(52):
	  #print(j)
	# get the prediction value for the first batch
    current_pred = lin_model.predict(current_batch)
    current_pred = scy.inverse_transform(current_pred)
	  #print(current_pred)
	# append the prediction into the array
    test_predictions.append(current_pred[0]) 
	  #print(test_predictions)
	# use the prediction to update the batch and remove the first value
    current_batch = np.array([np.append(current_batch[0,1:3],current_pred)])
	  #print(current_batch)
    current_batch.reshape(-1,1)

  pred2 = np.roll(test_predictions,-1)
  output2.append(pred2)
	#####################################################
```

```python
outputo2 = pd.DataFrame(output2)
outputo2.to_csv('SVR_Output.csv')
```
